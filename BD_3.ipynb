```python
pd.set_option("display.float_format", "{:.2f}".format)
data= pd.read_csv("/root/.cache/kagglehub/datasets/emanfatima2/spotify-global-hits-and-artist-analytics/versions/1/spotify_alltime_top100_songs.csv")
data.head()
```

```python
plt.figure(figsize=(12,6))
order = data['primary_genre'].value_counts().index

sns.countplot(data=data, x='primary_genre', order=order)
plt.xticks(rotation=90)
plt.title("Conteo de canciones por género")
plt.show()
```
